```python
from sklearn.model_selection import train_test_split #importing test train split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('generated_power_kw',axis=1),df['generated_power_kw'],test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test) #scaling the values so that values are in one array
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=7,weights='distance',p=1)
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_predict = knn.predict(x_test)
print(mean_squared_error(y_test,y_predict)) #getting the mean squared error
```

```python
r2 = r2_score(y_test,y_predict)
print("r2 score of model",r2) #good evaluation
```

```python
print("Adjacent R2 Error", (1 - ((1-r2)*(x_test.shape[0] -1)) / (x_test.shape[0]-1-1)))
```
